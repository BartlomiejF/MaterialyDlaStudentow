# W poniższej komórce można wpisać energię kwantu gamma w MeV

```python
hv = 0.66
```

#### Obliczenia i wizualizacja

```python
#  wygenerowanie zakresu kątów odbicia kwantu gamma od 0 do 2pi
fi = np.linspace(0,2*np.pi, 100)

#  obliczenie energii rozproszonego kwantu gamma
hv2 = hv/(1+(hv/0.511)*(1-np.cos(fi)))

#  wizualizacja
fig = plt.figure(figsize=(13,13))
ax = fig.add_subplot(221, projection='polar')
ax.plot(fi, (hv-hv2))
ax.set_title('electron energy')
ax = fig.add_subplot(222)
ax.plot(fi, (hv-hv2))
ax.set_title('electron energy')
ax.set_xlabel('fi')
ax.set_ylabel('E')
ax = fig.add_subplot(223, projection='polar')
ax.plot(fi, (hv2/hv))
ax.set_title('hv2/hv')
ax = fig.add_subplot(224)
ax.plot(fi, (hv2/hv))
ax.set_title('hv2/hv')
ax.set_xlabel('fi')
ax.set_ylabel('E')
plt.tight_layout()
```
