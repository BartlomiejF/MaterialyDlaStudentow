## W poniższej komórce można zmienić parametry rozkładów.
### Po zmianie należy nacisnąć ctrl + enter

```python
lambda_poisson = 3.5
lambda_gauss = 100
```

### Kliknij w poniższą komórkę i wciśnij ctrl + enter, a wyświetlą się wykresy

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

poi = np.random.poisson(lam=lambda_poisson, size=10000)
poi_gauss = np.random.poisson(lam=lambda_gauss, size=10000)
gx = np.unique(poi_gauss)
gaussx = np.linspace(gx.min(), gx.max(), 1000)
gaussy = norm.pdf(gaussx, lambda_gauss, np.sqrt(lambda_gauss))
fig, ax = plt.subplots(2,2, figsize=(15,12))
ax[0][0].hist(poi, density=True, bins=np.ptp(np.unique(poi)), alpha=0.6)
ax[0][0].set_title(f"lambda = {lambda_poisson}")
ax[0][0].axvline(x=np.mean(poi), c="r")
ax[0][0].axvline(x=np.median(poi), c="g", linestyle="--")
ax[0][1].hist(poi_gauss, density=True, bins=np.ptp(np.unique(poi_gauss)), alpha=0.6)
ax[0][1].set_title(f"lambda = {lambda_gauss}")
ax[0][1].axvline(x=np.mean(poi_gauss), c="r")
ax[0][1].axvline(x=np.median(poi_gauss), c="g", linestyle="--")
ax[0][1].plot(gaussx, gaussy, c="r")
ax[1][0].boxplot(poi, vert=False, medianprops=dict(linewidth=0))
ax[1][0].axvline(x=np.mean(poi), c="r")
ax[1][0].axvline(x=np.median(poi), c="g", linestyle="--")
ax[1][1].boxplot(poi_gauss, vert=False, medianprops=dict(linewidth=0))
ax[1][1].axvline(x=np.mean(poi_gauss), c="r")
ax[1][1].axvline(x=np.median(poi_gauss), c="g", linestyle="--")
plt.tight_layout()
```
